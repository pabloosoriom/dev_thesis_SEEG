```python
#Libraries
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Preprocessing 
df = pd.read_csv('mdb_output/T_File.csv', sep=';', encoding='UTF-8')
df['Patient_ID'] = df['Path_Local'].str.split('\\').str[1].str.split('_').str[1].astype(int)
#Making 2 new variables using the 2 variables after the spit, called Type and Format
df['Type'] = df['Path_Local'].str.split('\\').str[2].str.split('.').str[0]
df['Format'] = df['Path_Local'].str.split('\\').str[2].str.split('.').str[1]

#Function for taking patient information from T_File.csv
def get_patient_inf(df, patient_id):
    df=df[df['Patient_ID']==patient_id]
    df=df[['Patient_ID','Type','Format','Comment','Date']]
    df=df[df['Comment'].notnull()]
    df['Comment']=df['Comment'].fillna('No comment')
    return df
```
